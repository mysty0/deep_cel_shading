```python
def hex_to_rgb(value):
    value = value.lstrip('#')
    lv = len(value)
    return tuple(int(value[i:i + lv // 3], 16) for i in range(0, lv, lv // 3))


def rgb_to_hex(rgb):
    return '#%02x%02x%02x' % rgb

u, v, _ = hex_to_rgb("#e0dc00")

u = u/255
v = v/255
```

```python
def hex_to_rrgb(value):
    value = value.lstrip('#')
    lv = len(value)
    return tuple(float(int(value[i:i + lv // 3], 16))/255 for i in range(0, lv, lv // 3))

print(hex_to_rrgb("#b1b7ce"))
print(hex_to_rrgb("#ccb3a6"))
print(hex_to_rrgb("#eab4b3"))
print(hex_to_rrgb("#cfc9d9"))
print(hex_to_rrgb("#f3bfa9"))
print("==")
print(hex_to_rrgb("#f1bd9c"))
print(hex_to_rrgb("#d7a488"))
```
